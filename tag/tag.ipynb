```python
import pandas as pd
import numpy as np

df = pd.read_pickle('../data/comment_full.pkl')
df = df.loc[df.reply != '']
df['code'] = df[['comment','reply']].apply(lambda x: (x.comment[:5]+'_'+x.reply[:10]).replace('<br>','').replace('<br/>','').replace('\n',''), axis=1)
df = df.drop_duplicates(subset=['code']) 

df_table = pd.DataFrame()
nrow = 400
for i in range(round(np.ceil(len(df)/nrow))):
    for j in range(nrow):
        idx = nrow * i + j
        if idx < len(df):
            df_table.loc[j, f'code_{i}'] = df['code'].iloc[idx]
            df_table.loc[j, f'tag_{i}'] = 0
df_table.to_excel('table.xlsx')
```
